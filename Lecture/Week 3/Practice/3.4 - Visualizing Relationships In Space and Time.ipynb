# 3.4 - Visualizing Relationships In Space and Time

This notebook completes the practice activity using the built-in Gapminder dataset from Plotly Express.


## 1. Import libraries and load the Gapminder data

```python
import pandas as pd
import plotly.express as px

df = px.data.gapminder()
df.head()
```

## 2. Re-run the Gapminder choropleth map from lecture

This version shows **life expectancy** by country over time.


```python
df_year = df.sort_values('year')

fig_lifeexp = px.choropleth(
    df_year,
    locations='iso_alpha',
    color='lifeExp',
    hover_name='country',
    animation_frame='year',
    projection='natural earth',
    title='Life Expectancy by Country and Year',
    color_continuous_scale='Viridis'
)

fig_lifeexp.show()
```

## 3. Modify the `color=` attribute to show per capita GDP over time

```python
fig_gdp_color = px.choropleth(
    df_year,
    locations='iso_alpha',
    color='gdpPercap',
    hover_name='country',
    animation_frame='year',
    projection='natural earth',
    title='Per Capita GDP by Country and Year',
    color_continuous_scale='Viridis'
)

fig_gdp_color.show()
```

## 4. Change `color=` back to life expectancy and set `animation_frame=` to per capita GDP

To follow the instruction, the data is first sorted by `gdpPercap`.  
This plot technically works, but using **GDP as the animation frame** is not very effective because GDP has many unique values, so the animation is not as meaningful as using year.


```python
df_gdp = df.sort_values('gdpPercap').copy()
df_gdp['gdpPercap_str'] = df_gdp['gdpPercap'].round(2).astype(str)

fig_gdp_frame = px.choropleth(
    df_gdp,
    locations='iso_alpha',
    color='lifeExp',
    hover_name='country',
    animation_frame='gdpPercap_str',
    projection='natural earth',
    title='Life Expectancy by Country with Per Capita GDP as Animation Frame',
    color_continuous_scale='Viridis'
)

fig_gdp_frame.show()
```

## 5. Change `animation_frame=` back to year and change the color to population

The data is sorted by `year` before plotting.


```python
df_year = df.sort_values('year')

fig_population = px.choropleth(
    df_year,
    locations='iso_alpha',
    color='pop',
    hover_name='country',
    animation_frame='year',
    projection='natural earth',
    title='Population by Country and Year',
    color_continuous_scale='Viridis'
)

fig_population.show()
```

## 6. Reflection

The **best overall visualization** is the choropleth map animated by **year**, because time is naturally ordered and easy to interpret. Out of the three versions:

- **GDP as color over year** is useful for seeing economic differences between countries, but the values are highly skewed, so some countries dominate the scale.
- **Life expectancy as color with GDP as animation frame** is the least effective. GDP is not a natural time based sequence for animation, so the story is harder to follow.
- **Population as color over year** works better than GDP as frame, but it can still be harder to compare exact changes because very large countries dominate the color range.

Overall, the **life expectancy choropleth animated by year** is the clearest visualization for showing global trends over time. The year by year animation is intuitive, and life expectancy tends to produce a smoother, easier to read pattern across countries than GDP or population.
