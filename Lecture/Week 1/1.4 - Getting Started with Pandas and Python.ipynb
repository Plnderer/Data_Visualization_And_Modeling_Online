```python
import pandas as pd
```

```python
# 1. Read the titanic_passengers.csv file into a DataFrame
df = pd.read_csv("assets/titanic_passengers.csv")

# View the first few rows
df.head()
```

```python
# Check the columns
df.columns
```

```python
# 2. Create a subset including only passengers with 3rd class tickets
third_class = df[df["Pclass"] == 3]

# View the first few rows of the subset
third_class.head()
```

```python
# 3. Using .groupby(), calculate the mean age of those who survived and those who did not survive
mean_age_by_survival = third_class.groupby("Survived")["Age"].mean()

print(mean_age_by_survival)
```

```python
# 4. Create a new feature called 'Fare in CAD'
# 1 British pound = 1.70 Canadian dollars
df["Fare in CAD"] = df["Fare"] * 1.70

# View a few rows to confirm
df[["Fare", "Fare in CAD"]].head()
```

```python
# 5. Create a crosstab of passenger ticket class (Pclass) and Survived
crosstab_result = pd.crosstab(df["Pclass"], df["Survived"])

print(crosstab_result)
```

```python
# This version also adds row totals, which can make it easier to read
pd.crosstab(df["Pclass"], df["Survived"], margins=True)
```

```python
# 6. Create a new feature called 'Port_name' that recodes 'Embarked'
df["Port_name"] = df["Embarked"].map({
    "S": "Southampton",
    "C": "Cherbourg",
    "Q": "Queenstown"
})

# View a few rows to confirm
df[["Embarked", "Port_name"]].head(10)
```

```python
print("Mean age of 3rd class passengers by survival status:")
print(mean_age_by_survival)
print()

print("Crosstab of Pclass and Survived:")
print(crosstab_result)
print()

print("Preview of Fare in CAD and Port_name:")
print(df[["Fare", "Fare in CAD", "Embarked", "Port_name"]].head(10))
```
