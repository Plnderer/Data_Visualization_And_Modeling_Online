# 1.2 Data Visualization Exercise

This notebook loads the Titanic passenger dataset, previews the first 10 rows, and visualizes Fare, Survived, and Sex using clean tables and charts.

```python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display

pd.set_option('display.max_columns', None)
pd.set_option('display.width', 140)

# Read the CSV file into a DataFrame
df = pd.read_csv('assets/titanic_passengers.csv')

# Display the first 10 rows cleanly
print('First 10 rows of the DataFrame:')
display(df.head(10))
```

The `titanic_passengers.csv` file was successfully loaded into a Pandas DataFrame using `read_csv()`. The first 10 rows were displayed to confirm that the dataset loaded correctly and to preview the passenger information.

```python
# Histogram of Fare
plt.figure(figsize=(8,5))
plt.hist(df['Fare'].dropna(), bins=20, edgecolor='black')
plt.title('Histogram of Fare')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()

fare_summary = pd.DataFrame({
    'Statistic': ['Mean Fare', 'Median Fare', 'Minimum Fare', 'Maximum Fare'],
    'Value': [
        round(df['Fare'].mean(), 2),
        round(df['Fare'].median(), 2),
        round(df['Fare'].min(), 2),
        round(df['Fare'].max(), 2)
    ]
})

print('Fare Statistics:')
display(fare_summary)
```

The histogram of **Fare** shows a right-skewed distribution. Most passengers paid lower ticket prices, while a smaller number paid much higher fares. This suggests that the typical cost of passage was on the lower end. The summary table shows the average fare, median fare, and the full range of ticket prices in the dataset.

```python
# Bar chart of Survived
survived_counts = df['Survived'].value_counts().sort_index()

plt.figure(figsize=(6,4))
plt.bar(['Did Not Survive', 'Survived'], survived_counts.values, edgecolor='black')
plt.title('Bar Chart of Survival')
plt.xlabel('Survival Status')
plt.ylabel('Number of Passengers')
plt.show()

survival_rate = df['Survived'].mean()

survival_summary = pd.DataFrame({
    'Statistic': ['Proportion Survived', 'Percentage Survived'],
    'Value': [round(survival_rate, 4), f"{round(survival_rate * 100, 2)}%"]
})

print('Survival Statistics:')
display(survival_summary)
```

The bar chart of **Survived** shows the number of passengers who survived compared to those who did not survive. The summary table gives both the proportion and percentage of passengers who survived the voyage, making it easy to interpret the survival rate in the dataset.

```python
# Pie chart of Sex
sex_counts = df['Sex'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(sex_counts.values, labels=sex_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart of Passenger Sex')
plt.show()

sex_summary = pd.DataFrame({
    'Sex': ['male', 'female'],
    'Percentage': [
        f"{round((sex_counts.get('male', 0) / sex_counts.sum()) * 100, 2)}%",
        f"{round((sex_counts.get('female', 0) / sex_counts.sum()) * 100, 2)}%"
    ]
})

print('Sex Distribution:')
display(sex_summary)
```

The pie chart of **Sex** shows the proportion of male and female passengers in the dataset. The summary table provides the exact percentages for each group, showing that the dataset contains more male passengers than female passengers.
